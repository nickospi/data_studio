```python
raw_html = urlopen("https://www.azlyrics.com/b/bowie.html").read()
soup_doc = BeautifulSoup(raw_html, "html.parser")
```

```python
all_info = soup_doc.find_all(id='listAlbum') 
```

```python
bowietracks = []
for lines in all_info:
    urls = lines.find_all('a')
    for i in urls:
        url = i.get('href')
        bowietracks.append(url)
```

```python
print (bowietracks)
```

```python
df = pd.DataFrame(bowietracks)
```

```python
def get_all_lyrics(row):
    try:
        response = requests.get(row['0'])
        doc = BeautifulSoup(response.text, 'html.parser')

        title = doc.find(class_="biz-page-title").text
        lyrics = doc.find(class_="street-address").text
        album = doc.find(class_="i-stars")['title']

        return pd.Series({
            'title': title,
            'lyrics': lyrics,
            'album': album,

        })
    except:
        print('no')
df = df.apply(get_all_lyrics, axis=1).join(df)
df.head()
```
