### David Bowie was one of the most influential, long-lasting but also versatile artists of modern days. With a carreer spanning almost six decades he managed to remain active, current, constnalty reimagining himself.

```python
raw_html = urlopen("https://en.wikipedia.org/wiki/List_of_songs_recorded_by_David_Bowie").read()
soup_doc = BeautifulSoup(raw_html, "html.parser")
all_info = soup_doc.find_all('tr')[2:] 
```

```python
bowie=[]
for lines in all_info:
    data={}
    songs = lines.find_all('th')
    for song in songs:
        data['song_title'] = song(string=True)
    years = lines.find_all('td')[2:3]
    for i in years:
        data['year'] = i.find(string=True)
    albums = lines.find_all('td')[0:2]
    for album in albums:
        data['album_title'] = album.find(string=True)
    writers = lines.find_all('td')[0:1]
    for writer in writers:
        data['writer_name'] = writer.find(string=True)
        bowie.append(data)
```

```python
df=pd.DataFrame(bowie)
df=df.dropna()
```

```python
def print_sorted_vector(v):
    sorted_list = sorted(v.items(), key=lambda x: (x[1],x[0]), reverse=True) 
    sorted_list = sorted_list[:25]
    print('\n'.join([str(x) for x in sorted_list]))
```

```python
filepath = "NRC-Emotion-Lexicon-Wordlevel-v0.92.txt"
emolex_df = pd.read_csv(filepath,  names=["word", "emotion", "association"], sep='\t')
emolex_df = emolex_df.pivot(index='word', columns='emotion', values='association').reset_index()
```

```python
vec = TfidfVectorizer(vocabulary=emolex_df.word,
                      use_idf=False, 
                      norm='l1') 
matrix = vec.fit_transform(df_sixties['text'])
vocab = vec.get_feature_names()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)

wordcount_df = wordcount_df.iloc[:,1:]
```

```python
angry_words = emolex_df[emolex_df.anger == 1]['word']
anticipation_words = emolex_df[emolex_df.anticipation == 1]['word']
fear_words = emolex_df[emolex_df.fear == 1]['word']
negative_words = emolex_df[emolex_df.negative == 1]['word']
positive_words = emolex_df[emolex_df.positive == 1]['word']
sadness_words = emolex_df[emolex_df.sadness == 1]['word']
surprise_words = emolex_df[emolex_df.surprise == 1]['word']
```

```python
df_sixties['anger'] = wordcount_df[list(angry_words)].sum(axis=1)
df_sixties['positivity'] = wordcount_df[list(positive_words)].sum(axis=1)
df_sixties['fear'] = wordcount_df[list(fear_words)].sum(axis=1)
df_sixties['negative'] = wordcount_df[list(negative_words)].sum(axis=1)
df_sixties['sadness'] = wordcount_df[list(sadness_words)].sum(axis=1)
df_sixties['surprise'] = wordcount_df[list(surprise_words)].sum(axis=1)
df_sixties['anticipation'] = wordcount_df[list(anticipation_words)].sum(axis=1)
df_sixties[['anger','positivity','fear','negative','sadness','surprise','anticipation']].mean()
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111,polar=True)

sample = df_sixties[['anger','positivity','fear','negative','sadness','surprise','anticipation']].mean().tolist()
N = len(sample) 

theta = np.arange(0, 2*np.pi, 2*np.pi/N) 
bars = ax.bar(theta, sample, width=0.4)

ax.set_xticks(theta)
ax.set_xticklabels(['anger','positivity','fear','negativity','sadness','surprise','anticipation'])
ax.yaxis.grid(True)
ax.set_yticklabels([])


for r, bar in zip(theta, bars):
    bar.set_facecolor(plt.cm.viridis(r / 6.))
    bar.set_alpha(0.5)
plt.show()


#plt.savefig('2s.pdf', bbox_inches='tight')

plt.savefig('2s.png', bbox_inches='tight')
```

```python
df['text'] = df.text.str.replace('\n', ' ')
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['text'] = df['text'].str.replace('[^\w\s]','')
df['text'] = df['text'].str.replace('\syoure\s','')
df['text'] = df['text'].str.replace('\syouve\s','')
df['text'] = df['text'].str.replace('\sthing\s','')
df['text'] = df['text'].str.replace('\sthats\s','')
df['text'] = df['text'].str.replace('\szane\s','')
df['text'] = df['text'].str.replace('\sshes\s','')
df['text'] = df['text'].str.replace('\sarthur\s','')
df['text'] = df['text'].str.replace('\schingaling\s','')
df['text'] = df['text'].str.replace('\sduhdumduhdum\s','')
df['text'] = df['text'].str.replace('\sjanine\s','')
df['text'] = df['text'].str.replace('\swont\s','')
df['text'] = df['text'].str.replace('\syoull\s','')
df['text'] = df['text'].str.replace('\swasnt\s','')
df['text'] = df['text'].str.replace('\swhod\s','')
df['text'] = df['text'].str.replace('\swhoa\s','')
df['text'] = df['text'].str.replace('\sramona\s','')
df['text'] = df['text'].str.replace('\sthats\s','')
df['text'] = df['text'].str.replace('\sjulie\s','')
df['text'] = df['text'].str.replace('\sdont\s','')
df['text'] = df['text'].str.replace('\still\s','')
df['text'] = df['text'].str.replace('\scant\s','')
df['text'] = df['text'].str.replace('\sigive\s','')
df['text'] = df['text'].str.replace('\snabana\s','')
df['text'] = df['text'].str.replace('\sbro bro\s','')
df['text'] = df['text'].str.replace('\shabari\s','')
df['text'] = df['text'].str.replace('\shabari habari\s','')
df['text'] = df['text'].str.replace('\snabana nabana\s','')
df['text'] = df['text'].str.replace('\sasanti\s','')
df['text'] = df['text'].str.replace('\stata\s','')
df['text'] = df['text'].str.replace('\shee\s','')
df['text'] = df['text'].str.replace('\sjah\s','')
df['text'] = df['text'].str.replace('\sgoo\s','')
df['text'] = df['text'].str.replace('\sjah jah\s','')
df['text'] = df['text'].str.replace('\stvc\s','')











stop = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
freq = pd.Series(' '.join(df['text']).split()).value_counts()[:10]
freq = list(freq.index)
df['text']  = df['text'] .apply(lambda x: " ".join(x for x in x.split() if x not in freq))
freq = pd.Series(' '.join(df['text']).split()).value_counts()[-10:]
freq = list(freq.index)
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
df['text'] = df['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
df.year=df.year.astype(int)
```

```python
for i in df.year:
    if i < 1970:
        df.year = df.year.replace(i,'Sixties')
    if  1969 < i < 1980:
        df.year = df.year.replace(i,'Seventies')
    if  1979 < i < 1990:
        df.year = df.year.replace(i,'Eighties')
    if  1989 < i < 2000:
        df.year = df.year.replace(i,'Nineties')
    if  1999 < i < 2010:
        df.year = df.year.replace(i,'Zeroes')
    if  2009 < i < 2017:
        df.year = df.year.replace(i,'Final')

```

```python
nlp = spacy.load('en')
corpus = st.CorpusFromPandas(df, category_col='year', text_col='text',nlp=nlp).build()
```

```python
term_freq_df = corpus.get_term_freq_df()
term_freq_df['60sScore'] = \
corpus.get_scaled_f_scores('Sixties')
pprint(list(term_freq_df.sort_values(by='60sScore', ascending=False).index[:15]))
```

```python
term_freq_df = corpus.get_term_freq_df()
term_freq_df['70sScore'] = \
corpus.get_scaled_f_scores('Seventies')
pprint(list(term_freq_df.sort_values(by='70sScore', ascending=False).index[:15]))
```

```python
term_freq_df = corpus.get_term_freq_df()
term_freq_df['FinalScore'] = \
corpus.get_scaled_f_scores('Final')
pprint(list(term_freq_df.sort_values(by='FinalScore', ascending=False).index[:15]))
```
